```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets
%matplotlib notebook
n = 1000
```

```python
uniform_dist = np.random.uniform(-1, 1, n)
plt.plot(np.sort(uniform_dist))
```

```python
normal_dist = np.random.normal(0, 0.2, n)
plt.ylim(-1, 1)
plt.plot(np.sort(normal_dist))
```

```python
@ipywidgets.interact(sigma=(0.001,10.0,0.01))
def normal_family(sigma = 0.01):
    normal_dist = np.random.normal(0, sigma, n)
    plt.ylim(-1, 1)
    plt.plot(np.sort(normal_dist))
#ipywidgets.interact(normal_family)
```
