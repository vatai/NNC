```python
import sys
sys.path.append("../src")
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import Dense
from nnclib.model_dict import model_dict
model = model_dict['xception'][0]()
```

```python
def proc_weights(p, name=None):
    a = np.argsort(p, axis=0)
    p = np.take_along_axis(p, a, axis=0)
    norm = np.linalg.norm(p, axis=0)
    p /= norm[np.newaxis, :]
    plt.plot(p)
    if not name is None:
        plt.savefig(name)
    else:
        plt.show()


def process(model):
    # skip input layer
    for i, layer in enumerate(model.layers[1:]):
        print(type(layer))
        weights = layer.get_weights()
        shape = np.shape(weights)
        print(i, shape)
        if len(shape) > 1:
            sh1 = shape[-2]
            sh0 = shape[-1]
            for d in shape[:-2]:
                sh0 *= d
            weights = np.reshape(weights, [sh0, sh1])
            proc_weights(weights)


process(model)
```

```python
im = model.layers[32].get_weights()
shape = [np.shape(im)[-1], np.shape(im)[-2]]
im = np.reshape(im, shape)
```
