```python
%matplotlib inline
import matplotlib.pyplot as plt
from glob import glob
from os.path import join, basename
import numpy as np
from statsmodels.graphics.gofplots import qqplot
from scipy.stats import shapiro

base = "../report/weights/"

npy_names = glob(join(base, "*"))

results = {}
for npy_name in sorted(npy_names):
    layer = np.load(npy_name)
    print(npy_name)
    #print(np.shape(layer))
    table = []
    for idx, row in enumerate(layer):
        stat, pval = shapiro(row)
        var_result = np.var(row)
        table.append([stat, pval, var_result])
        #print('Shapiro: {}'.format(test_result))
        #print('Variance: {}'.format(var_result))
        #p = qqplot(row)
        #plt.show(p)
        #plt.close(p)
    
    model_name = basename(npy_name).split('_')[0]
    result = [basename(npy_name),np.array(table)]
    if model_name in results:
        results[model_name].append(result)
    else:
        results[model_name] =[result]
```

```python
for model_name, results in results.items():
    print('===== {} ====='.format(model_name))
    p = None
    for layer_name, result_tbl in results:
        print(layer_name)
        stat = result_tbl[:, 0]
        pval = result_tbl[:, 1]
        var = result_tbl[:, 2]
        mma = [pval.min(), pval.max(), pval.mean()]
        print("min: {}, max: {}, avg: {}".format(*mma))
        if p is None:
            p = plt.plot(np.sort(pval))
        else:
            p += plt.plot(np.sort(pval))
    plt.show(p)
```
