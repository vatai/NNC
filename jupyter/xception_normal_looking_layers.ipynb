```python
import sys
sys.path.append("../src")
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from keras.layers.convolutional import Conv2D
from keras.layers.core import Dense
from nnclib.utils import model_dic
model = model_dic['resnet50'][0]()
```

```python
def proc_weights(p):
    a = np.argsort(p, axis=1)
    p = np.take_along_axis(p, a, axis=1)
    norm = np.linalg.norm(p, axis=1)
    p /= norm[:, np.newaxis]
    plt.plot(p.T)
    plt.show()


def process(model):
    # skip input layer
    for layer in model.layers[1:]:
        print(type(layer))
        weights = layer.get_weights()
        if isinstance(weights, list) and weights:
            print('len', len(weights))
            weights = weights[0]
        shape = np.shape(weights)
        print(shape)
        if len(shape) > 1:
            sh1 = shape[-2]
            sh0 = shape[-1]
            for d in shape[:-2]:
                sh0 *= d
            weights = np.reshape(weights, [sh1, sh0])
            if isinstance(layer, Conv2D) or isinstance(layer, Dense):
                proc_weights(weights)


process(model)
```
